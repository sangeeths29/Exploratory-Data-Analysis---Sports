## Statistical Analysis of Players and Teams

```python
ManofMatch = eda1.groupby(['player_of_match']).count()['winner']
ManOfMatch_count = ManofMatch.sort_values(axis=0, ascending=False)
ManOfMatch_count.head()
```

```python
plt.figure(figsize=(8,6))
ax=sns.countplot(x='winner', data=eda1, order=pd.value_counts(eda1['winner']).index) 
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()
```

From the above two analysis, we can see that the most player of the match awards has been won by Chris Gayle and the team with most number of wins is Mumbai Indians.

Hence, we can conclude that the company should endorse the team Mumbai Indians and players like Chris Gayle who has won most number of man of the match awards and hence is a match winner.
